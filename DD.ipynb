```python
def outlier_detection_IQR(column): #checking the outliers in the data using IQR method.
    outliers = []
    Q1,Q3 = np.nanpercentile(column , [25,75]) 
    IQR = Q3 - Q1
    lowerbound = Q1 - (1.5 * IQR)
    upperbound = Q3 + (1.5 * IQR)
#The data points which fall below Q1 – 1.5 IQR or above Q3 + 1.5 IQR are outliers.
    for i in column: 
        if (i < lowerbound or i > upperbound):
            outliers.append(i)
    return outliers
```

```python
for item in numerical_variables.columns:
    outliers_sample = outlier_detection_IQR(data[item]) 
    print('outliers of variable with name', item, '==', outliers_sample) #return outliers.
```

```python
#treatment the outliers and extreme values
#Winsorization MethoD
for item in numerical_variables.columns:
    outliers_sample = outlier_detection_IQR(data[item]) 
    for i in data[item]:
        if i in outliers_sample:
            data.loc[(data[item] > np.nanpercentile(data[item], 80)), item] = np.nanpercentile(data[item], 80)
            data.loc[(data[item] < np.nanpercentile(data[item], 20)), item] = np.percentile(data[item], 20)
```

```python
#Our target variable depressed is extremely unblanced.
#This could lead to biased model performance.

#oversampling
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

# Assuming your dataset is stored in a DataFrame named 'data'
# X contains the features and y contains the target variable
X = Rdf.drop('depressed', axis=1)
y = Rdf['depressed']

# Initialize RandomOverSampler
oversampler = RandomOverSampler()

# Perform oversampling
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Convert oversampled data back to DataFrame if needed
oversampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['depressed'])], axis=1)
```
